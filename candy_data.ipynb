## 下载数据

```python
import pandas as pd
import numpy as np
```

```python
import requests

data_url = 'https://raw.githubusercontent.com/Praguexue/data/master/candy-power-ranking/candy-data.csv'

r = requests.get(data_url)

with open('candy_data.csv', 'wb') as f:
    f.write(r.content)
```

```python
data = pd.read_csv('candy_data.csv')
```

## 可视化

```python
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
data_sort_by_win = data.sort_values('winpercent', ascending=0)

data_sort_by_win.head(10)
```

我好像看到了**士力架**的身影

```python
data_sort_by_win.tail(10)
```

可怜的**水果糖**们

```python
win_30 = data_sort_by_win[:30]

win_30.shape
```

```python
type_list = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard']

win_types = win_30[type_list]
win_types.apply(np.sum, axis=0).sort_values(ascending=False).plot(kind='bar')
```

## <font color=red>最爱巧克力！</font>

```python
fail_30 = data_sort_by_win[-30:]

fail_30.shape
```

```python
fail_types = fail_30[type_list]
fail_types.apply(np.sum, axis=0).sort_values(ascending=False).plot(kind='bar')
```

## <font color=green>最讨厌水果硬糖！</font>

```python
data_sort_by_price = data.sort_values('pricepercent', ascending=0)

data_sort_by_price.head(10)
```

最贵的三种受欢迎度真的好低，有可能是价格的原因让人望而却步，也有可能是糖果不是受欢迎的类型，例如第2、3名就是非常不受欢迎的水果糖。

```python
price_30 = data_sort_by_price[:30]

price_30.shape
```

```python
price_type = price_30[type_list]
price_type.apply(np.sum, axis=0).sort_values(ascending=False).plot(kind='bar')
```

## <font color=green>巧克力的价格遥遥领先</font>

```python
cheap_30 = data_sort_by_price[-30:]

cheap_30.shape
```

```python
cheap_type = cheap_30[type_list]
cheap_type.apply(np.sum, axis=0).sort_values(ascending=False).plot(kind='bar')
```

## <font color=red>水果硬糖真是最便宜的</font>

最受欢迎的糖果里，我隐隐看到了士力架的身影：巧克力+花生+焦糖，不过价格也不低，不过现在大家生活水平都提升了，如果某种糖果具备了这三点，根据此篇（**并不具备概括性和全面性的**）分析，我猜测一样受市场欢迎。
