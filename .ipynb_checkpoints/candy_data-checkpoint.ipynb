# Candy_data  

## 下载数据

```python
import requests
import pandas as pd
import numpy as np
```

```python
data_url = 'https://raw.githubusercontent.com/Praguexue/data/master/candy-power-ranking/candy-data.csv'

r = requests.get(data_url)

with open('candy_data.csv', 'wb') as f:
    f.write(r.content)
```

```python
data = pd.read_csv('candy_data.csv')
```

## 可视化

```python
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
data_sort_by_win = data.sort_values('winpercent', ascending=0)

data_sort_by_win.head(10)
```

```python
win_50 = data_sort_by_win[:50]

win_50.shape
```

```python
win_50[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard']].sum().sort_values(ascending=False).plot(kind='bar')
```

## <font color=red>最爱巧克力！</font>

```python
fail_50 = data_sort_by_win[-50:]

fail_50.shape
```

```python
fail_50[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard']].sum().sort_values(ascending=False).plot(kind='bar')
```
